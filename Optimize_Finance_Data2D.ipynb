```python
##read csv file
import pandas as pd
import os

tikr = 'BAC'
pdate = 'Oct_19_2017'
path = tikr+'_'+pdate
f = os.listdir(path)

#weighting
mwt = .5 #weighting factor used to modify return attribute
pwt = 1-mwt #weighting factor used to modify last price attribute

#determine max and min values in contracts
mx = 0
mn = 5
lname = []

#iterate through all of the csv files in a folder to determine the maximum return value ('mx') and the minimum 
#contract price ('mn')
for i in range(len(f)):
    rd = pd.read_csv(path+'/'+f[i])
    rturn = rd.loc[:, 'return']
    price = rd.loc[:]['last price']
    if price.min() < mn:
        mn = price.min()
    if rturn.max() > mx:
        mx = rturn.max()
print(mn, mx)

#only consider returns > M_val (profitable contracts)
#assess all csv files in a folder based on a normalized weighting function
#the function is based on two attributes 'return' and 'last price'
#optimization function = 1 - ('return'/mx*mwt + mn/'last price'*pwt)
#if weighted equally (mwt = pwt) the optimization function would return 0 if one contract included both the maximum return
#and minimum price
#the function is designed to find the highest return at the lowest price within the file path
M_val = 1
trade = 1
con = 0
num = len(f)
idex = []

for i in range(len(f)):
    rd = pd.read_csv(path+'/'+f[i])
    rturn = rd.loc[:, 'return']
    price = rd.loc[:, 'last price']
    idex = []
    for j in range(len(rturn)):
        if rturn[j]>M_val and price[j]>0.05:
            idex = idex + [j]
    opt = rd.loc[idex][:]
    sm = opt.loc[:, 'return']/mx*mwt
    pv = mn/opt.loc[:, 'last price']*pwt
    #print(sm)
    o = 1-(sm+pv) # this is the optimization function
    #print(o.min(), trade)
    if o.min() < trade:
        trade = o.min()
        con = o.idxmin()
        num = i
rd = pd.read_csv(path+'/'+f[num])
print(trade)
print(f[num])
print(rd.loc[con, :])



```
