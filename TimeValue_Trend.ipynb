```python
#track a contract over time
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import pandas as pd
import os

def TimePlt(pdate,tikr):
    nmp = pd.DataFrame(data=None)
    ipath = 'Untitled Folder 1/Untitled Folder/'+tikr+'/Oct_'+str(pdate[0])+'_2017'
    f0 = os.listdir(ipath)[0]
    
    for i in range(len(pdate)):  #iterates through the days
        path = 'Untitled Folder 1/Untitled Folder/'+tikr+'/Oct_'+str(pdate[i])+'_2017'
        f = os.listdir(path)
        for j in range(len(f)):  #iterates through the contracts
            rd = pd.read_csv(path+'/'+f[j])
            #print(f[j])
            contract = rd.iloc[:, 0]
            r_val = rd.loc[:,'return']
            newdf = pd.concat([contract, r_val], axis = 1); #combine contracts and return values into one dataframe
        nmp = pd.concat([nmp,newdf],axis = 1); #update overal dataframe to include contracts from each day
    return nmp

```

```python
#Uses TimePlt function to collect stored csv file data and plot the results over the past days
tikr = 'BAC'
pdate = [14, 16, 17, 18, 19, 20, 23, 24]
ax = plt.subplot()

h = TimePlt(pdate,tikr)
for i in range(28):
    
    m = h.iloc[i,:]  #slices each of the contracts
    p = m.loc['return'] #extract return values
    
    if int(p.iloc[-1])>p.mean():
        ax.plot(pdate,p, 'y^-', label = m.iloc[0])
        continue

    ax.plot(pdate,p, 'go-', label = m.iloc[0])

#plt.scatter(pdate,p)

plt.title('Trend in Return')
plt.xlabel('date')
plt.ylabel('Return');
leg = ax.legend(loc=2, bbox_to_anchor=(1.05, 1.0))
```
